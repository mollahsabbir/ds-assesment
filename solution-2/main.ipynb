```python
import train
import model
import importlib
# importlib.reload(train)
# importlib.reload(model)

import torch
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, f1_score
```

```python
from train import train
from model import Conv2Model, Conv3Model, EnsembleModel
```

```python
results = train(Conv2Model())
```

```python
plt.plot(results["train_losses"], label='Training Loss')
plt.plot(results["val_losses"], label='Validation Loss')
plt.ylim(0,2)
plt.legend()
plt.show()
```

```python
plt.plot(results["train_accs"], label='Training Accuracy')
plt.plot(results["val_accs"], label='Validation Accuracy')
plt.ylim(0,100)
plt.legend()
plt.show()
```

```python
cm = confusion_matrix(results["y_true"], results["y_pred"])

classes = ['berry', 'bird', 'dog', 'flower']

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
f1_score(results["y_true"], results["y_pred"], average="micro")
```

# Train Ensemble Model

```python
conv2 = Conv2Model()
conv3 = Conv3Model()
```

```python
ensemble = EnsembleModel(conv2, conv3)
```

```python
ensemble_results = train(ensemble)
```

```python
plt.plot(ensemble_results["train_losses"], label='Training Loss')
plt.plot(ensemble_results["val_losses"], label='Validation Loss')
plt.ylim(0,2)
plt.legend()
plt.show()
```

```python
plt.plot(ensemble_results["train_accs"], label='Training Accuracy')
plt.plot(ensemble_results["val_accs"], label='Validation Accuracy')
plt.ylim(0,100)
plt.legend()
plt.show()
```

```python
cm = confusion_matrix(ensemble_results["y_true"], ensemble_results["y_pred"])

classes = ['berry', 'bird', 'dog', 'flower']

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
f1_score(ensemble_results["y_true"], ensemble_results["y_pred"], average="micro")
```
